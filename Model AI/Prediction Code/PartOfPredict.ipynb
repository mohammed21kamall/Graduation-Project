```python
# Define the prediction function using the provided equation
def predict_rating(ru, rv, rvi, Suv, degree_v, n_nearest_users, cu, cv):

    sum_weights = 0
    sum_rating_diff_weighted = 0

    for v in range(1, n_nearest_users + 1):
        sum_rating_diff_weighted += (rvi - rv) * Suv * w(cu, cv)  # Numerator
        sum_weights += (degree_v + 1) * abs(Suv)  # Denominator

    if sum_weights == 0:
        return 0  # Avoid division by zero

    predicted_rating = ru + (sum_rating_diff_weighted / sum_weights)
    return predicted_rating
```

```python
import pandas as pd

# Set display options to show all rows and columns without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Similarity
model_filename = 'drive/MyDrive/Dataset2/SimilarityOfUsers.csv'
Suv = pd.read_csv(model_filename)

# Clip values to 1.0
Suv = Suv.clip(upper=1.0)

# Apply lambda function to replace negative values with 0
Suv = Suv.applymap(lambda x: max(0, x))

# Apply lambda function to replace values less than 0.4 with 0.0
Suv = Suv.applymap(lambda x: x if x >= 0.4 else float(0.0))

# Display the DataFrame without truncation
print("DataFrame Suv:")
Suv.head(100)
```

```python
threshold=0.8
# Filter similar user IDs for a specified user ID
def get_similar_user_ids(user_id, threshold):
    # Filter the DataFrame to get similarity values for the specified user ID
    similarities_to_user = Suv[user_id]

    # Filter similar user IDs with similarity >= threshold
    similar_user_ids = similarities_to_user[similarities_to_user >= threshold].index.tolist()

    # Remove the specified user ID from the list if it's present
    similar_user_ids.remove(user_id) if user_id in similar_user_ids else None

    # Increase each user ID by 1
    similar_user_ids = [(int(user) + 1) for user in similar_user_ids]

    return similar_user_ids

# Prompt the user to enter the user ID
user_id = input("Enter the user ID: ")

# Check if the entered user ID exists in the DataFrame
if user_id not in Suv.columns:
    print("User ID not found!")
else:
    # Get similar user IDs for the specified user ID
    similar_user_ids = list(get_similar_user_ids(user_id, threshold))

    # Display similar user IDs
    print("Similar User IDs:", similar_user_ids)
```

```python

# 0.8 is the threshold i.e the cosine similarity should be at least greater than 0.8 between two users.
n = int(input("Enter the number of similar user IDs to display: "))
print("\nSimilarity Values: \n")
print("User ID" +" "+ "Similarity")
similarity_values = Suv[Suv[user_id] >= threshold][user_id].sort_values(ascending=False)[0:n]
print(similarity_values)
```

```python
# Function to get all movie IDs rated by users in a list
def get_list_rated_movie_ids(user_ids):
    rated_movies_by_users = {}  # Dictionary to store rated movies by each user

    for users in user_ids:
        # Filter ratings DataFrame for the specified user ID
        user_ratings = ratings[ratings['userId'] == users]

        # Extract and store the unique movie IDs rated by the user
        rated_movie_ids = user_ratings['movieId'].unique()
        rated_movies_by_users[users] = rated_movie_ids

    return rated_movies_by_users

# Example: Provide a list of user IDs
list_id = similar_user_ids  # Example list of user IDs

# Get all movie IDs rated by the specified users
rated_movies_by_users = get_list_rated_movie_ids(list_id)

# Print the movie IDs rated by each user
for users, rated_movies in rated_movies_by_users.items():
    print(f"Movie IDs rated by User {users}:")
    print(rated_movies)
    print("\n")
```

```python
movie_id_input = unrated_movies_list[2]
users_rated_movie = []
# Iterate over each user and their rated movies
for userid, rated_movies in rated_movies_by_users.items():
    # Check if the user has rated the input movie
    if movie_id_input in rated_movies:
        users_rated_movie.append(userid)

# Display the user IDs who have rated the input movie
print(f"User IDs who have rated movie {movie_id_input}:")
print(users_rated_movie)
```

```python
# Get the movie ID from unrated_movies_list
movie_id_input = unrated_movies_list[2]

# Initialize a dictionary to store similarity values with user IDs as keys
similarity_dict = {}

# Iterate over each user ID who has rated the input movie
for id in users_rated_movie:
    # Retrieve the similarity value between the user and the input user
    similarity_value = Suv.loc[id-1, user_id]
    # Add the similarity value to the dictionary with user ID as key
    similarity_dict[id] = similarity_value

# Display the similarity values along with the corresponding user IDs
print(f"\nSimilarity Values for Users who have rated movie {movie_id_input}:\n")
print(similarity_dict)
```

```python
n = int(input("Enter No.Of Elements : "))

# Sort the similarity_dict based on values in descending order
sorted_similarity_dict = dict(sorted(similarity_dict.items(), key=lambda item: item[1], reverse=True))

# Display the top n largest similarity values from the sorted dictionary
print(f"The {n} largest similarity values for Users who have rated movie {movie_id_input}:")
greatest_values = {k: sorted_similarity_dict[k] for k in list(sorted_similarity_dict)[:n]}
print(greatest_values)
```

```python
# Initialize empty lists to store keys and values
keys_list = []
values_list = []

# Iterate over the sorted similarity dictionary
for key, value in greatest_values.items():
    # Append keys to keys_list
    keys_list.append(key)
    # Append values to values_list
    values_list.append(value)

# Display the lists containing keys and values
print("Keys list:", keys_list)
print("Values list:", values_list)
```

```python
# Function to get rating for a user and movie
def get_rating(ratings, user_id, movie_id):
    rating = ratings[(ratings['userId'] == user_id) & (ratings['movieId'] == movie_id)]['rating'].values
    if len(rating) > 0:
        return rating[0]
    else:
        return None

# Prompt the user for input for two users and a movie
movie_id = movie_id_input
ratings_list = []

for i in range(n):
    id = keys_list[i]
    rating = get_rating(ratings, id, movie_id)
    ratings_list.append(rating)

# Print all ratings
print("Ratings for movie", movie_id, "are:", ratings_list)
```
