```python
import pandas as pd
import numpy as np
```

The Similarity (Suv)

```python

# Set display options to show all rows and columns without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Similarity
model_filename = 'drive/MyDrive/Dataset2/General Result/SimilartMatrix.csv'
Suv = pd.read_csv(model_filename)

# Clip values to 1.0
Suv = Suv.clip(upper=1.0)

# Apply lambda function to replace negative values with 0
Suv = Suv.applymap(lambda x: max(0, x))

# Apply lambda function to replace values less than 0.4 with 0.0
Suv = Suv.applymap(lambda x: x if x >= 0.4 else float(0.0))
Suv = Suv.drop(columns=['Unnamed: 0'])
# Display the DataFrame without truncation
print("DataFrame Suv:")
Suv.head()
```

```python
ratings = pd.read_csv('drive/MyDrive/Dataset2/General Result/ratings.csv')
ratings = ratings[['userId', 'movieId', 'rating']]
ratings.head()
```

```python
movie = pd.read_csv("drive/MyDrive/Dataset2/General Result/movies.csv")
movie = movie[['movieId', 'title']]
movie.head()
```

```python
# Extract the 'movieId' column and convert it into a list
movie_ids = movie['movieId'].tolist()

# Display the first few movie IDs
movie_ids[:10]  # Displaying the first 10 movie IDs as an example
```

```python
threshold=0.8
# Filter similar user IDs for a specified user ID
def get_similar_user_ids(user_id, threshold):
    # Filter the DataFrame to get similarity values for the specified user ID
    similarities_to_user = Suv[user_id]

    # Filter similar user IDs with similarity >= threshold
    similar_user_ids = similarities_to_user[similarities_to_user >= threshold].index.tolist()

    # Remove the specified user ID from the list if it's present
    similar_user_ids.remove(user_id) if user_id in similar_user_ids else None

    # Increase each user ID by 1
    similar_user_ids = [(int(user) + 1) for user in similar_user_ids]

    return similar_user_ids

# Prompt the user to enter the user ID
user_id = input("Enter the user ID: ")

# Check if the entered user ID exists in the DataFrame
if user_id not in Suv.columns:
    print("User ID not found!")
else:
    # Get similar user IDs for the specified user ID
    similar_user_ids = list(get_similar_user_ids(user_id, threshold))

    # Display similar user IDs
    print("Similar User IDs:", similar_user_ids)
```

```python
# Function to get all movie IDs rated by users in a list
def get_list_rated_movie_ids(user_ids):
    rated_movies_by_users = {}  # Dictionary to store rated movies by each user

    for users in user_ids:
        # Filter ratings DataFrame for the specified user ID
        user_ratings = ratings[ratings['userId'] == users]

        # Extract and store the unique movie IDs rated by the user
        rated_movie_ids = user_ratings['movieId'].unique()
        rated_movies_by_users[users] = rated_movie_ids

    return rated_movies_by_users

# Example: Provide a list of user IDs
list_id = similar_user_ids  # Example list of user IDs

# Get all movie IDs rated by the specified users
rated_movies_by_users = get_list_rated_movie_ids(list_id)

# Print the movie IDs rated by each user
for users, rated_movies in rated_movies_by_users.items():
    print(f"Movie IDs rated by User {users}:")
    print(rated_movies)
    print("\n")
```

```python
# List to store movie IDs not rated by User
unrated_movies_list = [movie_id for movie_id in movie_ids if movie_id not in rated_movies_list]

# Display the first few unrated movie IDs
print(f"Movie IDs not rated by User {user_id}:")
print(unrated_movies_list)  # Displaying the first 10 unrated movie IDs as an example
```

```python
movie_id_input = unrated_movies_list[2]
users_rated_movie = []
# Iterate over each user and their rated movies
for userid, rated_movies in rated_movies_by_users.items():
    # Check if the user has rated the input movie
    if movie_id_input in rated_movies:
        users_rated_movie.append(userid)

# Display the user IDs who have rated the input movie
print(f"User IDs who have rated movie {movie_id_input}:")
print(users_rated_movie)
```

```python
# Get the movie ID from unrated_movies_list
movie_id_input = unrated_movies_list[2]

# Initialize a dictionary to store similarity values with user IDs as keys
similarity_dict = {}

# Iterate over each user ID who has rated the input movie
for id in users_rated_movie:
    # Retrieve the similarity value between the user and the input user
    similarity_value = Suv.loc[id-1, user_id]
    # Add the similarity value to the dictionary with user ID as key
    similarity_dict[id] = similarity_value

# Display the similarity values along with the corresponding user IDs
print(f"\nSimilarity Values for Users who have rated movie {movie_id_input}:\n")
print(similarity_dict)
```

```python
n = int(input("Enter No.Of Elements : "))

# Sort the similarity_dict based on values in descending order
sorted_similarity_dict = dict(sorted(similarity_dict.items(), key=lambda item: item[1], reverse=True))

# Display the top n largest similarity values from the sorted dictionary
print(f"The {n} largest similarity values for Users who have rated movie {movie_id_input}:")
greatest_values = {k: sorted_similarity_dict[k] for k in list(sorted_similarity_dict)[:n]}
print(greatest_values)
```

```python
# Initialize empty lists to store keys and values
keys_list = []
suv = []

# Iterate over the sorted similarity dictionary
for key, value in greatest_values.items():
    # Append keys to keys_list
    keys_list.append(key)
    # Append values to values_list
    suv.append(value)

# Display the lists containing keys and values
print("Users ID:", keys_list)
print("Similairty Values:", suv)
```

```python
cluster_user = pd.read_csv("drive/MyDrive/Dataset2/General Result/NETWORK_CLUSTER.csv")
cluster_user = cluster_user[['Cluster Number', 'User IDs']]
cluster_user.head()
```

```python
# Initialize an empty dictionary to store the cluster number and user IDs
cluster_dict = {}

# Iterate over each row in the DataFrame
for index, row in cluster_user.iterrows():
    cluster_number = row['Cluster Number']
    user_ids = eval(row['User IDs'])  # Assuming the user IDs are stored as strings representing a list
    cluster_dict[cluster_number] = user_ids

# Print the dictionary
print(cluster_dict)
```

```python
# Function to search for a user and return its cluster number
def search_user(id):
    for cluster, users in cluster_dict.items():
        if id in users:
            return cluster
    return None  # Return None if user ID is not found in any cluster

# Example usage:
user_id_to_search = int(user_id)  # Replace "user123" with the user ID you want to search for
cluster_number_for_the_user = search_user(user_id_to_search)
if cluster_number_for_the_user is not None:
    print(f"The user {user_id_to_search} belongs to cluster {cluster_number_for_the_user}.")
else:
    print(f"The user {user_id_to_search} was not found in any cluster.")
```

```python
# Function to search for a user and return its cluster number
def search_list_user(user_ids):
    cluster_numbers = []
    for user_id in user_ids:
        for cluster, users in cluster_dict.items():
            if user_id in users:
                cluster_numbers.append(cluster)
                break  # Break the loop once the user is found in a cluster
        else:
            cluster_numbers.append(None)  # None if user not found in any cluster
    return cluster_numbers

# Example usage:
user_ids_to_search = [int(id) for id in keys_list]  # Replace user_ids with your list of user IDs
cluster_numbers = search_list_user(user_ids_to_search)

print(cluster_numbers)
```

```python
# Define the function w
def w(cu, cv):
    param1 = 1
    param2 = 0.7
    if cu == cv:
        return param1
    else:
        return param2

# Function to calculate weights between a user's cluster and a list of clusters
def calculate_weights(user_cluster, cluster_numbers):
    weights = []
    for cv in cluster_numbers:
        weight = w(user_cluster, cv)
        weights.append(weight)
    return weights

# Calculate weights
weights = calculate_weights(cluster_number_for_the_user, cluster_numbers)
print(weights)
```

```python
NoOfUsersInAllClusters = pd.read_csv("drive/MyDrive/Dataset2/NUMBER_OF_USER_IN_EACH_CLUSTER.csv")
NoOfUsersInAllClusters = NoOfUsersInAllClusters[['Cluster Number', 'Number of Users']]
NoOfUsersInAllClusters.head()
```

```python
# Function to search for the number of users in a specific cluster
def search_cluster(cluster_number):
    cluster_row = NoOfUsersInAllClusters[NoOfUsersInAllClusters['Cluster Number'] == cluster_number]
    if not cluster_row.empty:
        return cluster_row.iloc[0]['Number of Users']
    else:
        return None  # Return None if cluster number is not found

# Example usage:
cluster_number_to_search = cluster_number_for_the_user  # Replace 9 with the cluster number you want to search for
no_of_users = search_cluster(cluster_number_to_search)
if no_of_users is not None:
    print(f"Number of users in cluster {cluster_number_to_search}: {no_of_users}")
else:
    print(f"Cluster {cluster_number_to_search} not found.")
```

```python
# Function to search for the number of users in a specific cluster
def search_list_clusters(cluster_numbers):
    results = []
    for cluster_number in cluster_numbers:
        cluster_row = NoOfUsersInAllClusters[NoOfUsersInAllClusters['Cluster Number'] == cluster_number]
        if not cluster_row.empty:
            results.append(cluster_row.iloc[0]['Number of Users'])
        else:
            results.append(None)  # Append None if cluster number is not found
    return results

# Example usage:
no_of_cluster_list = search_list_clusters(cluster_numbers)

print(no_of_cluster_list)
```

```python
def degree(no_of_cluster1, no_of_cluster2):
    result =  no_of_cluster1 / no_of_cluster2
    return result

# Function to calculate weights between a user's cluster and a list of clusters
def calculate_degree(no_of_users_in_the_cluster_of_user, no_of_users_in_the_cluster_of_list):
    degrees = []
    for no_of_cluster2 in no_of_users_in_the_cluster_of_list:
        Degree = degree(no_of_users_in_the_cluster_of_user, no_of_cluster2)
        degrees.append(Degree)
    return degrees
# Calculate degrees
degrees = calculate_degree(no_of_users, no_of_cluster_list)
print(f"The Degree Of Users : {degrees}")
```

```python
def predict_rating(ru, rv_list, rvi_list, Suv_list, degree_v_list, n_nearest_users, w_list):
    sum_predicted_ratings = 0
    total_users = len(rv_list)
    absolute_errors = []

    for i in range(total_users):
        rv = rv_list[i]
        rvi = rvi_list[i]
        Suv = Suv_list[i]
        degree_v = degree_v_list[i]
        w = w_list[i]

        sum_weights = 0
        sum_rating_diff_weighted = 0

        for v in range(n_nearest_users):
            sum_rating_diff_weighted += (rvi - rv) * Suv * w  # Numerator
            sum_weights += (degree_v + 1) * abs(Suv)  # Denominator

        if sum_weights != 0:
            predicted_rating = ru + (sum_rating_diff_weighted / sum_weights)
            sum_predicted_ratings += predicted_rating

            # Calculate absolute error
            absolute_error = abs(predicted_rating - rv)
            absolute_errors.append(absolute_error)

    if total_users != 0:
        average_predicted_rating = sum_predicted_ratings / total_users
        mae = np.mean(absolute_errors)
        rmse = np.sqrt(np.mean(np.square(absolute_errors)))
        return mae, rmse, average_predicted_rating
    else:
        return None, None  # Return None if no users are provided
```

```python
mae, rmse, predictRating = predict_rating(ru, rv, rvi, suv, degrees, n, weights)
print(f"The Prediction Of User {user_id} For Movie No.of {movie_id_input} is : {predictRating}")
```
